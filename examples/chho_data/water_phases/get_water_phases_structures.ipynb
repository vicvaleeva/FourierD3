```python
import ase.io

# List of all XYZ files you previously created
mp_ids = [
    "mp-697111", "mp-1974803", "mp-703459", "mp-24043", "mp-696735",
    "mp-558226", "mp-557082", "mp-558958", "mp-697085", "mp-32959",
    "mp-2739191", "mp-634812", "mp-684704", "mp-1182019", "mp-673658",
    "mp-1181087", "mp-684678", "mp-990441",
]

all_atoms = []

# Read each XYZ back into memory
for mpid in mp_ids:
    infile = f"{mpid}.xyz"
    print(f"Reading {infile}...")
    ats = ase.io.read(infile)   # read single-frame xyz
    ats.info["comment"] = f"MP_ID={mpid}"
    all_atoms.append(ats)

# Write combined XYZ
outfile = "all_water_phases.xyz"
ase.io.write(outfile, all_atoms)

print(f"\nDone! Wrote {len(all_atoms)} structures into {outfile}")
```
